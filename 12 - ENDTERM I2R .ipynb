```r
#solve using matrix inversion method

#2x - y + 3z = 10
#x + 3y - 2z = 5
#3x - 2y + 4z = 12

#AX = B

A <- matrix(c(2,-1,3,1,3,-2,3,-2,4),nrow=3,ncol=3,byrow=T)
A_inv <- solve(A)                                            #solve function does the inverse calculation on matrix

B <- matrix(c(10,5,12),nrow=3,ncol=1)

A_inv%*%B #x=A_inv .B   
#x=2,y=3,z=3
```

```r
#Using LU decomposition

#we know AX=B -> LUX=B -> LY=B (#UX=Y) 

#first solve LY=B for Y

decompose <- lu.decomposition(A)
decom <- expand(decompose)
                                               
Y <- solve(decom$L)%*%B     

#solve UX =Y

x <- solve(decom$U)%*%Y 
x
```

```r
#Singular value decomposition 

smat <- matrix(sample(1:100,15),nrow=5,ncol=3)
smat

#A = U * D * t(V)

#computed by R

r_svd <- svd(smat)
u1 <- r_svd$u
d1 <- r_svd$d
v1 <- r_svd$v


#eigen computation

#find A.A^t  --->To find U

u <- smat%*%t(smat)

u_eigen <- eigen(u)
u_eigenvectors <- u_eigen$vectors
u_eigenvalues <- u_eigen$values


normalized_u_eigenvectors <- u_eigenvectors / sqrt(rowSums(u_eigenvectors^2))

U <- normalized_u_eigenvectors

#diagnol matrix using eigen values

diag_mat <- diag(c(u_eigenvalues),nrow =5,ncol=3)
diag_mat <- sqrt(diag_mat)

#find A^t.A  ---> To find V

v <- t(smat)%*%smat

v_eigen <- eigen(v)
v_eigenvectors <- v_eigen$vectors  # $ is used to address a specific column or a set of values using its title name
v_eigenvalues <- v_eigen$values    # [[]] can also be used instead of $


normalized_v_eigenvectors <- v_eigenvectors / sqrt(rowSums(v_eigenvectors^2))  #sqrt calculates the squareroot of the provided value

V_t <- t(normalized_v_eigenvectors)

# svd
svd <- U %*% diag_mat %*% V_t     # %*% is similar to dot product

round(smat,0) == round(svd,0)     # round function roundoff the values to a specified limit
```

```r
# Type of variable separation (continuous/categorical/disc_numerical)

variable_type<-function(dataframe,type) {
   
    continuous<-list()
    categorical<-list()
    disc_numerical<-list()
    
    for (col in colnames(dataframe)){
        if (class(dataframe[[col]])=='factor'){
            categorical<-append(categorical,col)            
        }
        else if (length(unique(dataframe[[col]]))>15){
            continuous<-append(continuous,col)            
        }else{
            disc_numerical<-append(disc_numerical,col)
        }
    }
    
    if (type=='continuous'){
        cat("Continuous variables:\n")
        return(unlist(continuous))
    }
    if (type=='categorical'){
        cat("\nCategorical variables:\n")
        return(unlist(categorical))
    }
    if (type=='disc_numerical'){
        cat("\nDiscrete numerical variables:\n")
        return (unlist(disc_numerical))
    }
}
```

```r
# Frequency Distribution using histogram & barplot

freq_plot<-function(dataframe){
    
    for (i in colnames(dataframe)){
        
        if (i %in% continuous){
            hist(dataframe[[i]],main=paste("Histogram of",i),xlab=i,ylab="Frequency",col="orange")
        }else{
            tbl <- with(dataframe, table(dataframe[i]))
            barplot(tbl,main=paste("Countplot of",i),xlab=i,ylab="Count",col="blue",beside = TRUE, legend = TRUE)
        }
        
    }
}
```

```r
#7. Create a function to shows values which are within the IQR Range

no_outliers = function(data){
  a=mean(data)-3*sd(data)     #lower bound
  b=mean(data)+3*sd(data)     #upper bound
  is.in=(data>a) & (data<b)
  print(sum(!is.in))          #no of outliers
  return(data[is.in])
}
```

```r
#8.write your own functions to compute the variance 

variance <- function(x, na.rm = TRUE) {
  n <- length(x)
  m <- mean(x, na.rm = TRUE)
  sq_err <- (x - m)^2
  sum(sq_err) / (n - 1)
}
```

```r
variance(1:10)
```

```r
#9.write your own functions to find skew of a numeric vector.

skewness <- function(x, na.rm = FALSE) {
  n <- length(x)
  m <- mean(x, na.rm = na.rm)
  v <- var(x, na.rm = na.rm)
  (sum((x - m) ^ 3) / (n - 2)) / v ^ (3 / 2)
}
```

```r
skewness(c(1, 2, 5, 100))
```

#11.Implement a fizzbuzz() function. It takes a single number as input. If the number is divisible by three, it returns “fizz”.
If it’s divisible by five it returns “buzz”. If it’s divisible by three and five, it returns “fizzbuzz”. Otherwise, it returns the number. Make sure you first write working code before you create the function.

```r
fizzbuzz <- function(x) {
  # these two lines check that x is a valid input
  stopifnot(length(x) == 1)
  stopifnot(is.numeric(x))
  if (!(x %% 3) && !(x %% 5)) {
    "fizzbuzz"
  } else if (!(x %% 3)) {
    "fizz"
  } else if (!(x %% 5)) {
    "buzz"
  } else {
    # ensure that the function returns a character vector
    as.character(x)
  }
}
```

```r
fizzbuzz(5)
fizzbuzz(3)
fizzbuzz(15)
```

```r
# 12.Write a R function for fibonnoci series
fibonacci <- function(n) {
  fib <- numeric(n)
  fib[1] <- 1
  fib[2] <- 1
  for (i in 3:n) {
    fib[i] <- fib[i - 1] + fib[i - 2]
  }
  return(fib)
}
```

```r
fibonacci(10)  #first 10 fibonnoci series
```

```r
#14.Compare two numeric vectors and count the no of times a number occurs
count_overlap <- function(vec1, vec2) {
  count <- 0
  for (x in vec1) {
    count <- count + sum(x == vec2)
  }
  return (count)
}
```

```r
vec1 <- c(1, 2, 3, 4, 5)
vec2 <- c(3, 4, 5, 6, 7)
count_overlap(vec1, vec2)
```

```r
#15.Create a moving average function with order 3

moving_average_order3 <- function(vec) {
  ma <- numeric(length(vec))
  for (i in 2:(length(vec) - 1)) {
    ma[i] <- mean(vec[(i-1):(i+1)])
  }
  return (ma)
}
```

```r
vec <- c(1:20)
moving_average_order3(vec)
```

```r
# Data Layer specifies the source of data
ggplot(data = mtcars)

# Aesthetic Layer (the way that data is mapped to visual elements in a plot)
ggplot(data = mtcars, aes(x = hp, y = mpg, col = disp))

# Geometric layer (Adding Size, color, and shape)
ggplot(data = mtcars, aes(x = hp, y = mpg, col = disp)) + geom_point() 

# Adding size
ggplot(data = mtcars,aes(x = hp, y = mpg, size = disp)) + geom_point()

# Adding color and shape
ggplot(data = mtcars,aes(x = hp, y = mpg, col = factor(cyl),shape = factor(am))) + geom_point()
```

```r
# Faceting
# (It is used to further drill down data and split the data by one or more variables,
# and then plot the subsets of the data altogether for optimum data visualization in R)

# split and visualize subsets of data
p <- ggplot(data = mtcars,
            aes(x = hp, y = mpg,
                shape = factor(cyl))) + geom_point()

# Separate rows according to transmission type
p + facet_grid(am ~ .)

# Separate columns according to cylinders
p + facet_grid(. ~ cyl)
```

```r
# Statistics layer
ggplot(data = mtcars, aes(x = hp, y = mpg)) +
  geom_point() +
  stat_smooth(method = lm, col = "red")
```

```r
# Coordinates layer: Control plot dimensions
ggplot(data = mtcars, aes(x = wt, y = mpg)) +
      geom_point() +
      stat_smooth(method = lm, col = "red") +
      scale_y_continuous("mpg", limits = c(2, 35),
                         expand = c(0, 0)) +
      scale_x_continuous("wt", limits = c(0, 25),
                         expand = c(0, 0)) + coord_equal()
```

```r
# Add coord_cartesian() to proper zoom in
ggplot(data = mtcars, aes(x = wt, y = hp, col = am)) +
      geom_point() + geom_smooth() +
      coord_cartesian(xlim = c(3, 6))
```

```r
# Theme layer: font size and background color
ggplot(data = mtcars, aes(x = hp, y = mpg)) +
       geom_point() + facet_grid(. ~ cyl) +
       theme(plot.background = element_rect(
       fill = "black", colour = "gray"))
```
